```python
class A:
    def __init__(self):
        print("Init A called!")
        self.a = 1
```

```python
class B(A):
    def __init__(self):
        super().__init__()
        print("Init B called!")
        self.b = self.a * 2
```

```python
class C(A):
    def __init__(self):
        super().__init__()
        print("Init C called!")
        self.c = self.a / 2
```

```python
class D(B, C):
    def __init__(self):
        super().__init__()
        print("Init D called!")
        self.d = self.b + self.c
```
