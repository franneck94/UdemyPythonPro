### Named Tuple

Tuple subclasses with named fields

- immutable
- hashable
- **\_\_repr\_\_**, **\_\_str\_\_** and compare defined
- Support indexing

```python
from collections import namedtuple
from pprint import pprint
```

```python
User = namedtuple("User", ["firstname", "lastname", "birthday"])
```

```python
user_jan = User(firstname="Jan", lastname="Schaffranek", birthday="24.02.1994")
user_peter = User(firstname="Peter", lastname="Peterson", birthday="01.01.1990")
```

```python
print(user_jan)
print(user_peter)
```

```python
print(user_jan.__repr__())
print(user_jan.__str__())
```

```python
print(user_jan.birthday)
print(user_jan.firstname)
print(user_jan.lastname)
```

```python
pprint(dir(namedtuple))
```
