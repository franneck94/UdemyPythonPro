```python
from dataclasses import dataclass
from dataclasses import field
```

### Standard Class


```python
class User:
    def __init__(self, name: str, age: int) -> None:
        self.name = name
        self.age = age
```

```python
u = User("jan", 27)

print(u)
```

```python
print(u.__dict__)
```

### DataClass (1)


```python
@dataclass
class User:
    name: str
    age: int
```

```python
u1 = User("jan", 27)

print(u1)
```

```python
print(u1.__dict__)
```

### DataClass (2)


```python
@dataclass(slots=True)
class User2:
    name: str
    age: int
```

```python
u2 = User2("jan", 27)

print(u2)
```

```python
# print(u2.__dict__)
```

### DataClass (3)


```python
@dataclass(slots=True)
class User3:
    name: str
    age: int
    is_active: bool = False
    orders: list[float] = field(
        default_factory=list,
        compare=False,
        hash=False,
        repr=False,
    )

    def method(self):
        print(self.name)
```

```python
u3 = User3("jan", 27)

print(u3)

u3.method()
```
