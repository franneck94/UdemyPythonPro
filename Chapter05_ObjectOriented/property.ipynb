### Properties


```python
from functools import cached_property
```

```python
class User1:
    def __init__(self, name: str, user_id: int) -> None:
        self._name = name
        self._user_id = user_id

    def __repr__(self) -> str:
        return f"{self._name}, {self._user_id}"

    def get_name(self) -> str:
        return self._name

    def set_name(self, new_name: str) -> None:
        self._name = new_name

    def del_name(self) -> None:
        self._name = ""
```

```python
class User2:
    def __init__(self, name: str, user_id: int) -> None:
        self._name = name
        self._user_id = user_id

    def __repr__(self) -> str:
        return f"{self._name}, {self._user_id}"

    @property
    def name(self) -> str:
        return self._name

    @name.setter
    def name(self, new_name: str) -> None:
        self._name = new_name

    @name.deleter
    def name(self) -> None:
        self._name = ""
```

```python
u1 = User2("Jan", 123)
print(u1)
print(u1.name)
u1.name = "Peter"
print(u1)
```

```python
class User:
    def __init__(self, name, user_id):
        self._name = name
        self._user_id = user_id

    def __repr__(self):
        return f"{self._name}, {self._user_id}"

    @cached_property
    def name(self):
        return self._name
```

```python
u1 = User("Jan", 123)
print(u1)
u1.name = "Peter"
print(u1)
```
