```python
from pprint import pprint
from typing import Any
from typing import Callable
from typing import Optional
```

## Functions


#### Default Arguments and Return Values


```python
def g(
    p1: int,
    p2: int,
    p3: int = 0,
    p4: int = 10,
) -> None:
    print(f"p1: {p1}, p2: {p2}, p3: {p3}, p4: {p4}")
```

```python
in1 = 10
in2 = 20
in3 = 30
in4 = 40

g(
    p1=in1,
    p2=in2,
    p3=in3,
    p4=in4,
)
```

```python
def append_list(
    my_list: list[int],
    value: int,
) -> None:
    my_list.append(value)
```

```python
lst = [1, 2]

lst = append_list(lst, 3)
print(lst)
```

#### Function Attributes


```python
def grow_list(
    val: int,
    my_list: Optional[list[int]] = None,
) -> list[int]:
    if my_list:
        my_list.append(val)
    else:
        my_list = [val]
    return my_list
```

```python
my_func = grow_list

print(my_func, type(my_func))
```

```python
def print_function_output(
    fn: Callable,
    **kwargs: Any,
) -> None:
    print(fn(**kwargs))
```

```python
print_function_output(my_func, val=10)
```

```python
my_func(10, [1, 2])
```

```python
pprint(dir(grow_list))
```
