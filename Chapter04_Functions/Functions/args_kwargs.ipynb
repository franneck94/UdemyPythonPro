```python
from typing import Any
from typing import Optional
```

```python
def my_function(
    a: int,
    b: int,
    c: int = 10,
    d: int = 12,
    e: int = 13,
) -> None:
    print(f"a: {a}, b: {b}, c: {c}, d: {d}, e:{e}")
```

```python
my_function(1, 2)
```

### Positional Arguments


\*args: variable number of pos. arguments


```python
def my_function2(
    a: Any,
    b: Any,
    *args: Any,
) -> None:
    print(*args, type(args))
    print(f"a: {a}, b: {b}, args: {args}")
```

```python
my_function2(1, 2, 3, 4)
my_function2(1, 2, 3)
my_function2(1, 2)
```

```python
def my_function3(*args: Any) -> None:
    print(*args, type(args))
    print(f"args: {args}")
```

```python
my_function2(1, 2, 3, 4)
```

NORMAL ARGS; \*ARGS; DEFAULT ARGS


```python
def my_function4(
    a: Any,
    *args: Any,
    b: Optional[Any] = None,
) -> None:
    print(*args, type(args))
    print(f"a: {a}, b: {b}, args: {args}")


# my_function4(1, b=2, 3, 4) # errors
```

### Keyword Arguments


-   Normal Args; \*Args; Default Args; \*\*Kwargs
-   \*Args: Tuple
-   \*\*Kwargs: Dict


```python
def my_function(
    a: Any,
    *args: Any,
    x: int = 2,
    y: int = 3,
    z: int = 4,
    **kwargs: Any,
) -> None:
    print(args, type(args))
    print(kwargs, type(kwargs))
    print(f"a: {a}, x: {x}, y: {y} , z: {z}\nargs: {args}\nkwargs: {kwargs}")
```

```python
my_function(
    1,
    3,
    4,
    x=1337,
    b=False,
    c=30,
    d=40.5,
)
```

```python
def my_function(
    *args: Any,
    **kwargs: Any,
) -> None:
    print(f"args: {args}\nkwargs: {kwargs}")
```

```python
my_function(
    1,
    3,
    4,
    x=13.37,
    b=False,
    c=30,
    d=40.5,
)
```
