```python
import time
from threading import Thread
```

# Threads


```python
def worker(sleep_time: float) -> None:
    print("Start worker")

    time.sleep(sleep_time)

    print("End worker")
```

```python
t1 = Thread(target=worker, name="t1", args=(2.0,))
```

```python
print(f"Ident: {t1.ident}")

print(f"Alive: {t1.is_alive()}")

print(f"Name: {t1.name}")
```

```python
t1.start()


print(f"Ident: {t1.ident}")

print(f"Alive: {t1.is_alive()}")

print(f"Name: {t1.name}")


t1.join()
```
